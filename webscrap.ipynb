```python
import requests
from bs4 import BeautifulSoup
import csv
import  pandas as pd
```

```python
url="https://www.flipkart.com/search?q=samsung+mobiles"
r=requests.get(url)
soap=BeautifulSoup(r.content,"html.parser")
tit=soap.find_all('div',class_='_4rR01T')
pric=soap.find_all('div',class_='_30jeq3 _1_WHN1')
rev=soap.find_all('div',class_='')
print(pric)
```

```python
titt=[]
pri=[]
for tittle ,price in zip(tit,pric):
    titt.append(tittle.text)
    pri.append(price.text)
print(titt)
```

```python
data={'tittle':titt,'price':pri}
data
```

```python
model=pd.DataFrame(data=data)
model
```
